```python
df3_copy['Carrera'] = df3_copy['Carrera'].apply(unidecode)
df3_copy.head()
```

Aplicamos unidecode para hacer mas legibles los nombres de las carreras en la columna carreras

Este código elimina los acentos y caracteres especiales de los nombres de carrera en df3_copy, usando la función unidecode. Esto permite estandarizar el texto y facilita la comparación entre carreras que podrían tener diferencias por tildes o caracteres especiales.

```python
def remove(text):
  if not isinstance(text, str):
    return text  # Return as is if not a string
  new_text = ""
  prev_char = ""
  for char in text:
    if char.lower() == 'a' and prev_char.lower() == 'a':
      continue
    new_text += char
    prev_char = char
  return new_text

df3_copy['Carrera'] = df3_copy['Carrera'].apply(remove)

df3_copy.head()
```

Funcion solo para eliminar letras a repetidas debido a que el data frame original presenta alteraciones y letras repetidas sobretodo la que mencinada anteriormente

define una función de Python llamada remove que toma un argumento, text.
La primera línea comprueba si la entrada text no es una cadena utilizando isinstance(text, str). Si no es una cadena, la función devuelve inmediatamente la entrada original text sin realizar ningún cambio. Esto evita errores si la función se llama con tipos de datos distintos de cadenas.

Si la entrada es una cadena, inicializa una cadena vacía llamada new_text que almacenará la cadena modificada. También inicializa una cadena vacía prev_char para mantener un registro del carácter anterior procesado.

Luego, el código itera a través de cada carácter en la entrada text utilizando un bucle for.
Dentro del bucle, comprueba si el carácter actual (char) es una 'a' (sin distinguir mayúsculas y minúsculas, usando .lower()) y si el carácter anterior (prev_char) también era una 'a' (sin distinguir mayúsculas y minúsculas).

Si ambas condiciones son verdaderas (lo que significa que hay 'a' consecutivas), se ejecuta la instrucción continue. Esto salta el resto de la iteración actual y pasa al siguiente carácter en la cadena de entrada, ignorando efectivamente la 'a' actual.
Si los caracteres no son 'a' consecutivas, el carácter actual (char) se agrega a la cadena new_text.

Finalmente, la variable prev_char se actualiza al carácter actual (char) para la siguiente iteración.
Después de procesar todos los caracteres, la función devuelve la cadena new_text, que es la cadena original con cualquier carácter 'a' consecutivo eliminado.

df3_copy['Carrera'] selecciona la columna llamada 'Carrera' del DataFrame df3_copy.

.apply(remove) es un método de pandas que aplica una función a cada elemento de una Serie (que es una columna de un DataFrame). En este caso, aplica la función remove a cada valor de cadena en la columna 'Carrera'.

El resultado de esta operación (las cadenas modificadas con las 'a' consecutivas eliminadas) se asigna de nuevo a la columna 'Carrera' en el DataFrame df3_copy, sobrescribiendo los valores originales.

●	¿Cómo ha evolucionado el total de matriculados en educación superior desde 2017 hasta 2024?

```python
primer_df = df2_limpio[['AÑO', 'MATRICULADOS MUJERES POR PROGRAMA', 'MATRICULADOS HOMBRES POR PROGRAMA' ,'TOTAL MATRICULADOS', "CLASIFICACIÓN INSTITUCIÓN NIVEL 1",'CLASIFICACIÓN INSTITUCIÓN NIVEL 2' ,'CLASIFICACIÓN INSTITUCIÓN NIVEL 3', "NOMBRE CARRERA" ]].copy()
primer_df.head()
```

```python
# Filtrar por el rango de años deseado
primer_df['AÑO'] = pd.to_numeric(primer_df['AÑO'])
df_filtrado = primer_df[(primer_df['AÑO'] >= 2017) & (primer_df['AÑO'] <= 2024)].copy()
# Agrupar por año y sumar el total de matriculados
df_agrupado_año = df_filtrado.groupby('AÑO')['TOTAL MATRICULADOS'].sum().reset_index()
# Ordenar por año para el gráfico
df_agrupado_año = df_agrupado_año.sort_values('AÑO')
# Graficar la evolución del total de matriculados por año
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_agrupado_año, x='AÑO', y='TOTAL MATRICULADOS')
plt.title('Evolución del Total de Matriculados por Año (2017-2024)')
plt.xlabel('Año')
plt.ylabel('Total de Matriculados')
plt.grid(True)
plt.xticks(df_agrupado_año['AÑO']) # Asegura que se muestren todos los años en el eje x
plt.tight_layout()
plt.show()

df_agrupado_año.head(8)
```

```python
# Calcular la suma total de matriculados en df5
total_matriculados_2025 = df5['Matriculados'].sum().copy()
# Crear una nueva fila para df_agrupado_año
nueva_fila = pd.DataFrame({'AÑO': [2025], 'TOTAL MATRICULADOS': [total_matriculados_2025]})
# Concatenar la nueva fila al DataFrame df_agrupado_año
df_agrupado_año = pd.concat([df_agrupado_año, nueva_fila], ignore_index=True)

#grafico
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_agrupado_año, x='AÑO', y='TOTAL MATRICULADOS', marker='o')
plt.title('Evolución del Total de Matriculados por Año (2017-2025)')
plt.xlabel('Año')
plt.ylabel('Total de Matriculados')
plt.grid(True)
plt.xticks(df_agrupado_año['AÑO'])
plt.tight_layout()
plt.show()
```

##Pregunta 1

●	¿Qué áreas del conocimiento han experimentado mayor crecimiento en matrícula durante el período analizado?

```python
# Calcular la matrícula total por área del conocimiento para cada año
df_area_año = df2_limpio.groupby(['AÑO', 'ÁREA DEL CONOCIMIENTO'])['TOTAL MATRICULADOS'].sum().reset_index()
# Calcular la matrícula en el año inicial y final para cada área
año_inicial = df_area_año['AÑO'].min()
año_final = df_area_año['AÑO'].max()
matricula_inicial = df_area_año[df_area_año['AÑO'] == año_inicial].set_index('ÁREA DEL CONOCIMIENTO')['TOTAL MATRICULADOS']
matricula_final = df_area_año[df_area_año['AÑO'] == año_final].set_index('ÁREA DEL CONOCIMIENTO')['TOTAL MATRICULADOS']
# Calcular el crecimiento absoluto y porcentual
# Usar fill_value=0 para áreas que no existen en ambos años
crecimiento_absoluto = matricula_final.sub(matricula_inicial, fill_value=0)
crecimiento_porcentual = ((matricula_final.sub(matricula_inicial, fill_value=0)) / matricula_inicial.replace(0, np.nan)).fillna(0) * 100 # Evitar división por cero
# Crear un DataFrame con los resultados
df_crecimiento_areas = pd.DataFrame({
    'Matrícula Inicial': matricula_inicial,
    'Matrícula Final': matricula_final,
    'Crecimiento Absoluto': crecimiento_absoluto,
    'Crecimiento Porcentual': crecimiento_porcentual
}).fillna(0) # Rellenar NaN con 0 si un área solo aparece en un año
# Ordenar por crecimiento absoluto y porcentual
df_crecimiento_absoluto_sorted = df_crecimiento_areas.sort_values('Crecimiento Absoluto', ascending=False)
df_crecimiento_porcentual_sorted = df_crecimiento_areas.sort_values('Crecimiento Porcentual', ascending=False)
```

```python
df_crecimiento_porcentual_sorted.head(10)
```

```python
# Visualizar el crecimiento absoluto de las top N áreas
top_n = 10
plt.figure(figsize=(12, 8))
sns.barplot(x=df_crecimiento_absoluto_sorted.head(top_n).index, y=df_crecimiento_absoluto_sorted['Crecimiento Absoluto'].head(top_n), palette='viridis')
plt.title(f'Top {top_n} Áreas del Conocimiento con Mayor Crecimiento Absoluto de Matrícula ({año_inicial}-{año_final})')
plt.xlabel('Área del Conocimiento')
plt.ylabel('Crecimiento Absoluto de Matriculados')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
# Visualizar el crecimiento porcentual de las top N áreas (considerando áreas con matrícula inicial > 0)
df_crecimiento_porcentual_filtered = df_crecimiento_areas[df_crecimiento_areas['Matrícula Inicial'] > 0].sort_values('Crecimiento Porcentual', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=df_crecimiento_porcentual_filtered.head(top_n).index, y=df_crecimiento_porcentual_filtered['Crecimiento Porcentual'].head(top_n), palette='magma')
plt.title(f'Top {top_n} Áreas del Conocimiento con Mayor Crecimiento Porcentual de Matrícula ({año_inicial}-{año_final}) (Áreas con Matrícula Inicial > 0)')
plt.xlabel('Área del Conocimiento')
plt.ylabel('Crecimiento Porcentual de Matriculados')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
